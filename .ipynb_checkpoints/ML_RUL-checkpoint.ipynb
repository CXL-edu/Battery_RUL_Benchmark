## Define random seeds and process data.

```python
""" Define random seeds and process data. """
import random
import numpy as np
import pandas as pd
import sklearn

import matplotlib.pyplot as plt
import matplotlib.animation as animation


""" Define random seeds. """
random_seed = 42
random.seed(random_seed)
np.random.seed(random_seed)
sklearn.utils.check_random_state(random_seed)

""" Define data source. """
data_path_dic = {
    'NASA': 'data/NASA/',
    'CALCE': 'data/CALCE/',
}

data_group_list_dic = {
    'NASA': ['data'],   # ['B0005', 'B0006', 'B0007', 'B0018']
    'CALCE': ['data'],
}



class DataProcess(object):
    """
    Processing data for fixed-length input and output predictions.

    Parameters
    ----------
    val_flag : bool, default=False
        When using libraries such as scikit-learn, their models are trained using
        cross-validation without the need for additional validation sets.

    individual : bool, default=False
        If false, then use the front part of each battery's complete lifecycle data as the training set,
        and the data from the later part of the lifecycle as the test and validation sets.

        If true, then use complete data from some batteries as the training set,
        while the complete data from other batteries will be used as separate test and validation sets.
        
    """
    def __init__(self, val_flag=False, individual=True):
        self.val_flag = val_flag   # Whether to use independent validation set.
        self.individual = individual    # Choose different training set and test set partitioning modes.


    def train_test_split(self, data:pd.DataFrame, test_ratio=0.2):
        if not self.individual:
            # Convert type of data to list[list[float]] from pd.DataFrame.
            data_temp = []
            for i in range(data.shape[1]):
                data_temp.append(data.iloc[:, i].dropna().values.tolist())
            data = data_temp

            if not self.val_flag:
                train_x, train_y = [], []
                test_x, test_y = [], []

                train_data, test_data = data[:-int(len(data)/4)], data[-int(len(data)/4):]
                for i in range(len(train_data)):
                    for j in range(len(train_data[i])-seq_len-pred_len+1):
                        train_x.append(train_data[i][j:j+seq_len])
                        train_y.append(train_data[i][j+seq_len:j+seq_len+pred_len])
                for i in range(len(test_data)):
                    for j in range(len(test_data[i])-seq_len-pred_len+1):
                        test_x.append(test_data[i][j:j+seq_len])
                        test_y.append(test_data[i][j+seq_len:j+seq_len+pred_len])
                return train_x, train_y, test_x, test_y

            else:
                train_x, train_y = [], []
                val_x, val_y = [], []
                test_x, test_y = [], []
                ...
                return train_x, train_y, val_x, val_y, test_x, test_y
        else:
            if not self.val_flag:
                pass
            else:
                pass


    def process_data_2_RUL(self, data_name:str, data_group:str, seq_len=seq_len, pred_len=pred_len, val_flag=None, individual=None):
        self.val_flag = val_flag if val_flag is not None else self.val_flag
        self.individual = individual if individual is not None else self.individual
        data_load_method = getattr(self, 'load_' + data_name + '_data')
        data = data_load_method(data_path_dic[data_name], data_group)

        return self.train_test_split(data)


    def load_NASA_data(self, data_path, data_group):
        data = pd.read_csv(data_path + data_group + '.csv')
        return data

    def load_CALCE_data(self, data_path, data_group):
        pass


""" Define hyper-parameters. """
seq_len, pred_len = 10, 8

data_loader = DataProcess()
data_source = 'NASA'
train_x, train_y, test_x, test_y = data_loader.process_data_2_RUL(data_source, data_group_list_dic[data_source][0],
                                                                  seq_len=seq_len, pred_len=pred_len, val_flag=False, individual=False)
```

## Visualize the dynamic process of using LR iteration to predict capacity.
using NASA data as an example.

```python
""" Define a dynamic graph using LR for iterative prediction. """
from sklearn.linear_model import LinearRegression

real_capacity = pd.read_csv('data/NASA/data.csv').iloc[:, -1].dropna().values.tolist()
model = LinearRegression().fit(train_x, train_y)


def iterative_pred(model, input, EOL):
    # Iterative prediction
    input_list, pred_y = [input], []
    while True:
        pred_y.append(model.predict([input])[0])
        if all(pred_y[-1] > EOL):
            input = np.concatenate([input[len(pred_y[-1]):], pred_y[-1]])
            input_list.append(input)
        else:
            break
    return input_list, pred_y

class visual_iterative_pred():
    def __init__(self, model, input, start, EOL, real_capacity, ax):
        self.input_list, self.pred_y = iterative_pred(model, input, EOL)
        self.in_len, self.out_len, self.s = len(input), len(self.pred_y[0]), start
        capacity, = ax.plot(real_capacity, color='k', linewidth=2, markersize=3, marker='o', linestyle='dashed', label='Capacity')
        raw_input, = ax.plot(np.arange(self.s, self.s+len(input)), input, color='b', linewidth=2, marker='*', markersize=6, label='raw_input')
        EOL_im = ax.axhline(y=EOL, color='r', linestyle='dashed', linewidth=2, label='EOL')
        self.input_im, = ax.plot([], [], color='r', linewidth=2, marker='o', markersize=6, label='iter_input')
        self.predicted, = ax.plot([], [], color='green', linewidth=3, markersize=12, label='predicted')
        self.pred_im, = ax.plot(np.arange(self.s+self.in_len, self.out_len + self.s+self.in_len), self.pred_y[0],
                                color='lime', linewidth=2, marker='*', markersize=4, label='predicting')
        ax.legend()
        ax.set_xlabel('Cycle')
        ax.set_ylabel('Capacity(Ahr)')
        self.ims = [capacity, raw_input, EOL_im, self.pred_im, self.input_im]

    def my_update(self, i):
        self.input_im.set_data(np.arange(self.s+i*8, self.s+i*8+self.in_len), self.input_list[i])
        self.pred_im.set_data(np.arange(self.s+i * self.out_len + self.in_len, self.s+(i + 1) * self.out_len + self.in_len),
                              self.pred_y[i])
        if i > 0:
            self.predicted.set_data(np.arange(self.s+self.in_len, self.s+i*self.out_len+self.in_len), np.concatenate(self.pred_y[:i]))

        return self.ims


fig, ax = plt.subplots(figsize=(7, 4), dpi=300)
start_cycle = 25
scope = visual_iterative_pred(model, test_x[start_cycle], start_cycle, 1.4, real_capacity, ax)

# pass a generator in "emitter" to produce data for the update func
ani = animation.FuncAnimation(fig, scope.my_update, range(len(scope.pred_y)), interval=250, blit=True)
ani.save(f'example_LR_NASA_{start_cycle}.gif', writer='pillow')
```
